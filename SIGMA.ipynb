```python
import glob
import pandas as pd

txtnames = []
for file in glob.glob("*.txt"):
    txtnames.append(file)
    
txtnames
```

```python
files={}
for filename in txtnames:
    with open(filename,"r") as file:
        if filename in files:
            continue
        files[filename]=file.read()
        
type(files)
```

```python
df=pd.DataFrame.from_dict(files, orient='index', columns=['Contenuto'])
df=df.rename_axis('Nome').reset_index()
df.head()
```

## conteggio frequenze relative di ogni parola nella propria frase

```python
from collections import Counter

# series of dictionaries of word counts, each series entry covering one row
N_d = df['Contenuto'].str.split().apply(Counter)
print(N_d)
```

```python
def calc_rel_freq_sent(counter):
    sum_w_sent= sum(counter.values())
    rel={}
    for key in counter:
        rel[key]=round((counter[key]/sum_w_sent),4)
    return rel 

freq_rel_w_sent= N_d.apply(calc_rel_freq_sent)
print(freq_rel_w_sent)
```

```python
prova=freq_rel_w_sent.values.tolist()
```

```python
df_freq_rel_w_sent=pd.DataFrame(prova)
df_freq_rel_w_sent=df_freq_rel_w_sent.T
df_freq_rel_w_sent=df_freq_rel_w_sent.rename_axis('parola').reset_index()
```

```python
df_freq_rel_w_sent= df_freq_rel_w_sent.sort_values('parola').reset_index(drop=True)
df_freq_rel_w_sent.rename(columns = df["Nome"], inplace = True)    
df_freq_rel_w_sent.head()
```

## conteggio frequenze relative di ogni parola sul totale documenti

```python
#freq parole sul corpus
N_D=df.Contenuto.str.split(expand=True).stack().value_counts()
sum_w_doc=sum(N_D)
freq_rel_w_doc=round((N_D/sum_w_doc),4)
freq_rel_w_doc
```

```python
freq_rel_w_doc=freq_rel_w_doc.sort_index() 
#freq_rel_w_doc
```

```python
df_freq_rel_w_doc=freq_rel_w_doc.to_frame()
df_freq_rel_w_doc=df_freq_rel_w_doc.rename_axis('parola').reset_index()
df_freq_rel_w_doc.rename(columns = {0:'Freq rel doc'}, inplace = True) 
df_freq_rel_w_doc.head()
```

```python
freq=pd.merge(df_freq_rel_w_sent, df_freq_rel_w_doc, on='parola')
freq=freq.fillna(0)
freq.head()
```

## sigma

```python
for column in freq.iloc[:,1:17]:
    freq['sigma' + str(column)]= (freq[column]-freq['Freq rel doc'])/freq['Freq rel doc']
freq=freq.fillna(0)
freq.head()
```

```python
freq.to_excel("Frequenze Relative Parole_NO Numeri.xlsx") 
```

```python
print(freq.columns)
```

```python
newstuff=[]

for column in freq.iloc[:,18:34]:
    newstuff.append(pd.DataFrame(data=(freq.nlargest(3, column))))
newstuff
```

# BIGRAMS

```python
import pandas as pd
from collections import Counter
import nltk
from nltk.util import ngrams 
import nltk
```

```python
N_d2 = df['Contenuto'].apply(lambda row: Counter(ngrams(row.split(), 2)))
print(N_d2)
```

```python
def calc_rel_freq_sent(counter):
    sum_w_sent= sum(counter.values())
    rel={}
    for key in counter:
        rel[key]=round((counter[key]/sum_w_sent),4)
    return rel 

freq_rel_bi_sent= N_d2.apply(calc_rel_freq_sent)
print(freq_rel_bi_sent)
```

```python
prova=freq_rel_bi_sent.values.tolist()
```

```python
df_freq_rel_bi_sent=pd.DataFrame(prova)
df_freq_rel_bi_sent=df_freq_rel_bi_sent.T
df_freq_rel_bi_sent=df_freq_rel_bi_sent.rename_axis('bigrams').reset_index()
df_freq_rel_bi_sent.head()
```

```python
df_freq_rel_bi_sent= df_freq_rel_bi_sent.sort_values('bigrams').reset_index(drop=True)
df_freq_rel_bi_sent.rename(columns = df["Nome"], inplace = True)    
df_freq_rel_bi_sent.head()
```

### conteggio frequenze relative di ogni bigram sul totale documenti

```python
ngram_list = [pair for row in df['Contenuto'] for pair in ngrams(row.split(), 2)]
freq_bigrams_doc = Counter(ngram_list).most_common()
N_D2 = pd.DataFrame(freq_bigrams_doc, columns=['bigrams', 'count'])
N_D2.head()
```

```python
sum_bi_doc=sum(N_D2['count'])
N_D2['freq_rel_bi_doc']=round((N_D2['count']/sum_bi_doc),4)
N_D2.head()
```

```python
N_D2.sort_values(by=['bigrams'], inplace=True)
N_D2.head()
```

```python
freq2=pd.merge(df_freq_rel_bi_sent, N_D2, on='bigrams')
freq2.drop('count', inplace=True, axis=1)
freq2=freq2.fillna(0)
freq2.head()
```

## sigma

```python
for column in freq2.iloc[:,1:17]:
    freq2['sigma' + str(column)]= (freq2[column]-freq2['freq_rel_bi_doc'])/freq2['freq_rel_bi_doc']
freq2.head()
```

```python
freq2.to_excel("Frequenze Relative Bigrams_NO Numeri.xlsx") 
```

```python
newstuff=[]

for column in freq2.iloc[:,18:34]:
    newstuff.append(pd.DataFrame(data=(freq2.nlargest(3, column))))
newstuff
```
